```python
## Basic T-Distribution
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np

## Setup      
dof = 10        # Degrees of freedom
dof2 = 50
dof3 = 1000
alpha = 0.05    # Significence level
ntails = 2      # Number of tails 

plt.figure()
xs = np.linspace(-5,5,1000)
plt.plot(xs, st.t.pdf(xs,dof), 'k', label="t-dist df="+str(dof))
# plt.plot(xs, st.norm.pdf(xs,dof), 'b', label="z-dist df="+str(dof))

plt.plot(xs, st.t.pdf(xs,dof2), 'r', label="t-dist df="+str(dof2))
plt.plot(xs, st.t.pdf(xs,dof3), 'g', label="t-dist df="+str(dof3))
plt.legend()
```
