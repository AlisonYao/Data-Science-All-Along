```python
import matplotlib.pyplot as plt 
import numpy as np
```

# Generate Data

```python
rng = np.random.RandomState(42)
x = 10 * rng.rand(50) # 50 random x's
print('x:', x, len(x))
y= 2 * x - 1 + rng.randn(50) # 50 random error -> 50 random y's
print('y:', y, len(y))
plt.scatter(x, y)
```

# Choose a class of model

A **class** of model is not the same as an **instance** of a model.

```python
from sklearn.linear_model import LinearRegression
```

# Choose model hyperparameters

* Would we like to fit for the offset (i.e., intercept)?
* Would we like the model to be normalized?
* Would we like to preprocess our features to add model flexibility? 
* What degree of regularization would we like to use in our model? 
* How many model components would we like to use?

```python
model = LinearRegression(fit_intercept=True)
model # now it's an instance
```

# Arrange data into a features matrix and target vector

```python
X = x[:, np.newaxis]
X.shape
```

```python
X
```

```python
# y is alreaady in the right format
y
```

# Fit the model to your data

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

# Predict labels for unknown data

For the sake of this example, our “new data” will be a grid of x values, and we will ask what y values the model predicts:

```python
xfit = np.linspace(-1, 11) # Return evenly spaced numbers over a specified interval. default: 50 points
xfit, len(xfit)
```

```python
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
yfit
```

```python
plt.scatter(xfit, yfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit) # plotting 折线, this happens to be a straight line
```
